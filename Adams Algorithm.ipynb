```python
import pandas as pd
import numpy as np
import os
import math
dataFileName = os.getcwd() + "/Desktop/AssnData.csv"
dataset = pd.read_csv(dataFileName)
def logisticOutput(w,x,b):
    return 1/(1 + np.exp(-((w * x) + b)))

def deltaW(y,w,x,b):
    sigmoidOutput = logisticOutput(w,x,b)
    return ((sigmoidOutput - y) * sigmoidOutput * (1 - sigmoidOutput) * x)

def deltaB(y,w,x,b):
    sigmoidOutput = logisticOutput(w,x,b)
    return ((sigmoidOutput - y) * sigmoidOutput * (1 - sigmoidOutput))

def loss(X,Y,w,b):
    error = 0
    for x,y in zip(X,Y):
        error += (0.5 * (math.pow((y - logisticOutput(w,x,b)) , 2)))
    return error    

def AdamsAlgorithm(X,Y):
    eta,w,b,epoch,beta1,beta2 = 0.01,1,1,100,0.9,0.99
    mw,mb,vw,vb = 0,0,0,0
    epsilon = 1e-8
    
    bmw,bmb = 0,0
    bvw,bvb = 0,0
    
    
    for i in range(1,epoch+1):
        dw,db = 0,0
        for x,y in zip(X,Y):
            dw += deltaW(y,w,x,b)
            db += deltaB(y,w,x,b)
        mw = (beta1 * mw) + ((1 - beta1)*dw)
        mb = (beta1 * mb) + ((1 - beta1)*db)
        
        vw = (beta2 * vw) + ((1 - beta2) * (dw*dw))
        vb = (beta2 * vw) + ((1 - beta2) * (db*db))
        
        bmw = mw / (1 - math.pow(beta1,i))
        bmb = mb / (1 - math.pow(beta1,i))
        
        bvw = vw / (1 - math.pow(beta2,i))
        bvb = vb / (1 - math.pow(beta2,i))
        
        w = w - ((eta / np.sqrt(bvw + epsilon)) * bmw)
        b = b - ((eta / np.sqrt(bvb + epsilon)) * bmb)
        print "Loss after " , i , " iterations : "  ,  loss(X,Y,w,b)
    return (w,b)

def __Main():
    X = dataset['X']
    Y = dataset['Y']
    w,b = AdamsAlgorithm(X,Y)
    print "Loss after 100 iterations : " , loss(X,Y,w,b)
    

__Main()
```
