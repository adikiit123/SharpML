```
import sklearn.datasets
from sklearn.metrics import confusion_matrix
import numpy as np
from keras.optimizers import Adam
from keras.layers import Dense,Dropout
from keras.models import Sequential
from keras.losses import binary_crossentropy
from keras.utils import np_utils

dataset = sklearn.datasets.load_breast_cancer()

inputDataSize = len(dataset.data[:])
trainingSampleCount = (int)(inputDataSize * 0.7)
randomIndexArray = np.random.permutation(inputDataSize)

# Test,Train Split
xTrainingArray = dataset.data[randomIndexArray[:trainingSampleCount]]
yTrainingArray = dataset.target[randomIndexArray[:trainingSampleCount]]

xTestArray = dataset.data[randomIndexArray[trainingSampleCount:]]
yTestArray = dataset.target[randomIndexArray[trainingSampleCount:]]

cancerPredictorModel = Sequential()

hiddenLayer1 = Dense(30,activation = 'relu')
hiddenLayer2 = Dense(15,activation = 'relu')
outputLayer = Dense(1,activation = 'sigmoid')

# Preparing Model
cancerPredictorModel.add(hiddenLayer1)
cancerPredictorModel.add(Dropout(0.3))
cancerPredictorModel.add(hiddenLayer2)
cancerPredictorModel.add(Dropout(0.3))
cancerPredictorModel.add(outputLayer)

cancerPredictorModel.compile(optimizer = Adam(),loss = binary_crossentropy,metrics = ['accuracy'])
cancerPredictorModel.fit(xTrainingArray,yTrainingArray,10,150)

(loss,metric) = cancerPredictorModel.evaluate(xTestArray,yTestArray,10,50)

cancerPredictorModel.save('./breastCancerModel.h5')
yPred = cancerPredictorModel.predict(xTestArray)
yPred = [1 if y > 0.5 else 0 for y in yPred]
matrix = confusion_matrix(yTestArray, yPred)
print(matrix)
```
