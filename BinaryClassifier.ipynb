```python
# DEPENDENCIES
import numpy as np
import tensorflow as tf
from sklearn import datasets

# LOADING DATASET AND TENSOR DECLARATION
# 2 HIDDEN LAYERS (4 NEURONS EACH)
# 1 OUTPUT LAYER (1 NEURON DETERMINING THE CLASS) 

dataSet = datasets.load_iris()

# SUPERVISED LEARNING PARAMS
numberOfClasses = 3
numberOfFeatures = 4
numberOfEpochs = 1000
learningRate = 0.1

# FETCHING THE DATA
XData = dataSet.data
YData = dataSet.target

trainingSampleSize = (int)((XData.size / numberOfFeatures)* 0.8)
randomArray = np.random.permutation(trainingSampleSize)

xTrainingData = XData[randomArray]
yTrainingData = YData[randomArray]

#PLACEHOLDERS FOR INPUT,HIDDEN AND OUTPUT LAYER NEURONS DEFINED.
# Define the number of hidden layers an the
# number of neurons for each layer
nHidden1 = 4
nHidden2 = 4
nOutput = 1

# Inputs and outputs
x = tf.placeholder(tf.float32,[None, numberOfFeatures])
y = tf.placeholder(tf.float32,[None,1])
  
# Model
def multilayer_perceptron(x, weights, biases):
    # Hidden layer with RELU activations
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.relu(layer_1)
    # Hidden layer with sigmoid activations
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    # Output layer with linear activations
    out_layer = tf.matmul(layer_2, weights['out']) + biases['out_b']
    return out_layer
 
# define the weights and the biases for each layer
 
weights = {
    'h1': tf.Variable(tf.truncated_normal([numberOfFeatures, nHidden1]),name="h1"),
    'h2': tf.Variable(tf.truncated_normal([nHidden1, nHidden2]),name="h2"),
    'out': tf.Variable(tf.truncated_normal([nHidden2, nOutput]),name="out"),
    }
biases = {
    'b1': tf.Variable(tf.truncated_normal([nHidden1]),name="b1"),
    'b2': tf.Variable(tf.truncated_normal([nHidden2]),name="b2"),
    'out_b': tf.Variable(tf.truncated_normal([nOutput]),name="out_b"),
    }
 
# Initialization
init = tf.global_variables_initializer()
 
saver = tf.train.Saver()
 
# Call your model defined
yy = multilayer_perceptron(x, weights, biases)
 
# Define the cost function and optimizer
cost_function = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=yy, labels=y))
training_step = tf.train.GradientDescentOptimizer(learningRate).minimize(cost_function)

# Launch the graph
sess = tf.Session()
sess.run(init)

mse_history = []
accuracy_history = []

# Calculate the cost and the accuracy for each epoch
for epoch in range(numberOfEpochs):
    sess.run(training_step, feed_dict={x:xTrainingData, y:yTrainingData})
    cost = sess.run(cost_function,feed_dict={x:xTrainingData, y:yTrainingData})
    cost_history = np.append(cost_history, cost)
    correct_prediction = tf.equal(tf.argmax(yy,1),tf.argmax(y,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))    
    print('epoch: ', epoch,' - ', 'cost: ', cost, "- Train Accuracy: ", accuracy)
    
```
